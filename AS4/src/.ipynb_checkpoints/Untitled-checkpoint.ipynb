```python
import argparse
import cv2
import numpy as np
import os
import glob
import pandas as pd
def mouse_click(event, x, y, flags, param):
    global mouse_point, img
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),4,(255,255,255),-1)
        cv2.imshow('img',img)
        mouse_point.append([x, y])
```

```python
CHECKERBOARD = (6,9)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)


objpoints = []
imgpoints = []

objp = np.zeros((1, CHECKERBOARD[0] * CHECKERBOARD[1], 3), np.float32)
objp[0,:,:2] = np.mgrid[0:CHECKERBOARD[0], 0:CHECKERBOARD[1]].T.reshape(-1, 2)
prev_img_shape = None
cv2.namedWindow('img')
images = glob.glob('./images/*.jpg')
for fname in images:
    mouse_point = []
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    ret, corners = cv2.findChessboardCorners(gray, CHECKERBOARD,
                                             cv2.CALIB_CB_ADAPTIVE_THRESH +
                                             cv2.CALIB_CB_FAST_CHECK +
                                             cv2.CALIB_CB_NORMALIZE_IMAGE)
    
    
    if ret == True:
        objpoints.append(objp) 
        
        corners2 = cv2.cornerSubPix(gray, corners, (11,11),(-1,-1), criteria)
        
        imgpoints.append(corners2)
        img = cv2.drawChessboardCorners(img, CHECKERBOARD, corners2, ret)
    
    with open("./points"+fname[8:-3]+"txt", 'w') as file:
        for i in range(len(imgpoints[0])):
            s = " ".join(map(str, objpoints[0][0][i])) + " " + " ".join(map(str, imgpoints[0][i]))[1:-1]
            file.write(s+'\n')
    
    cv2.imshow('img',img)
    k = cv2.waitKey(0)
    while(k == ord('m')):
        cv2.setMouseCallback("img", mouse_click)
        h = cv2.waitKey(0)
        if h == ord('q'):
            with open("./points"+fname[8:-3]+"txt", 'w') as file:
                for i in range(len(mouse_point)):
                    s = " ".join(map(str, objpoints[0][0][i])) + " " + " ".join(map(str, mouse_point[i]))[1:-1]
                    file.write(s+'\n')
            break
    
                
            
    
cv2.destroyAllWindows()

"""h,w = img.shape[:2]
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)

print("Camera matrix : \n")
print(mtx)
print("dist : \n")
print(dist)
print("rvecs : \n")
print(rvecs)
print("tvecs : \n")
print(tvecs)"""
```

```python
points = pd.read_csv("./points/left01.txt", delimiter='\s+', header = None, dtype = np.float64)
objpoints_file = points[[0, 1, 2]]
imgpoints_file = points[[3, 4]]
A = np.zeros((2*points.shape[0], 9))
h = np.zeros((9, 1))
j = 0
for i in range(0, points.shape[0], 2):
    A[i][0] = objpoints_file[0][j]
    A[i][1] = objpoints_file[1][j]
    A[i][2] = objpoints_file[2][j]
    A[i][6] = objpoints_file[0][j] * -1 * imgpoints_file[3][j]
    A[i][7] = objpoints_file[1][j] * -1 * imgpoints_file[3][j]
    A[i][8] = objpoints_file[2][j] * -1 * imgpoints_file[3][j]
    A[i+1][3] = objpoints_file[0][j]
    A[i+1][4] = objpoints_file[1][j]
    A[i+1][5] = objpoints_file[2][j]
    A[i+1][6] = objpoints_file[0][j] * -1 * imgpoints_file[4][j]
    A[i+1][7] = objpoints_file[1][j] * -1 * imgpoints_file[4][j]
    A[i+1][8] = objpoints_file[2][j] * -1 * imgpoints_file[4][j]
    j += 1
u, d, v = np.linalg.svd(A)
v[-1]
```
